```python
import pandas as pd

Id_document = '1qfYRfl1Udw9GuspPncVJXG21ouqn-y02NYuuuepukUs'

sheet_sales = 'sales_in_Paraguay'
sheet_distributors = 'distributors_profiles'
sheet_exports = 'exports_to_Paraguay'
sheet_locations = 'locations_profiles'

URL = f'https://docs.google.com/spreadsheets/d/{Id_document}/gviz/tq?tqx=out:csv&sheet='
```

```python
df_sales = pd.read_csv(URL + sheet_sales)
df_sales = df_sales.drop(df_sales.index[-1])

display(df_sales)
```

```python
columns_to_convert = ['Rubber', 'Brass', 'Vinyl', 'Granite', 'Stone', 'Brick', 'Aluminum', 'Glass', 'Plexiglass', 'Steel', 'Wood', 'Plastic']

# for column in columns_to_convert:
#     df_sales[column] = df_sales[column].astype(str)
#     df_sales[column] = df_sales[column].str.replace(',', '').astype(float)

df_sales[columns_to_convert] = df_sales[columns_to_convert].astype(float)

display(df_sales)

```

```python
totals_sales = df_sales[['Rubber', 'Brass', 'Vinyl', 'Granite', 'Stone', 'Brick',
				   'Aluminum', 'Glass', 'Plexiglass', 'Steel', 'Wood', 'Plastic']].sum()
print(totals_sales)
```

```python
df_distributors = pd.read_csv(URL + sheet_distributors)
display(df_distributors)
```

```python
df_exports = pd.read_csv(URL + sheet_exports)
df_exports[columns_to_convert] = df_exports[columns_to_convert].astype(float)
display(df_exports)
```

```python
totals_exports = df_exports[['Rubber', 'Brass', 'Vinyl', 'Granite', 'Stone', 'Brick',
				   'Aluminum', 'Glass', 'Plexiglass', 'Steel', 'Wood', 'Plastic']].sum()
print(totals_exports)
```

```python
df_locations = pd.read_csv(URL + sheet_locations)
display(df_locations)
```

```python
def inspect_df(df, name):
    print(f"Inspeccionando {name}:")
    print("\n")
    print(f"Shape: {df.shape}")
    print("\n")
    print(f"Columns: {df.columns}")
    print("\n")
    print(f"Data types:\n{df.dtypes}")
    print("\n")
    print(f"Null values:\n{df.isnull().sum()}")
    print("\n")
    print("**" * 30)
    print("\n")
```

```python
inspect_df(df_sales, 'df_sales')
inspect_df(df_distributors, 'df_distributors')
inspect_df(df_exports, 'df_exports')
inspect_df(df_locations, 'df_locations')
```

```python
df_sales.drop_duplicates(inplace=True)
df_distributors.drop_duplicates(inplace=True)
df_exports.drop_duplicates(inplace=True)
df_locations.drop_duplicates(inplace=True)
```

```python
print ('Sales:')
print (totals_sales)
print("\n")
print("Exports:")
print (totals_exports)
print("\n")

combined_totals = totals_sales - totals_exports
print("Combined totals:")
print(combined_totals.astype(float).round(2))
```

```python
combined_totals_df = combined_totals.to_frame(name='Total')
combined_totals_df.loc['Total of All Categories'] = combined_totals_df.sum()

print(combined_totals_df.astype(int))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Plotting
ax = combined_totals_df.plot(kind='bar', figsize=(14, 7), color='blue')  # Initial plot

# Title and labels
plt.title('Combined Totals of Categories')
plt.xlabel('Category')
plt.ylabel('Total')
plt.xticks(rotation=45)

# Mark the zero line
plt.axhline(0, color='grey')

# Change color of bars below zero to red
for p in ax.patches:  # Loop through all the bars
	if p.get_height() < 0:  # If the bar is below zero
		p.set_color('red')  # Change the color to red
		
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

## Análisis detallado:

### Discrepancias significativas:

La mayoría de las categorías muestran una diferencia positiva entre ventas y exportaciones, lo que sugiere que se están vendiendo más productos en Paraguay de los que se están exportando desde CMM.
Sin embargo, hay dos categorías que muestran discrepancias negativas muy grandes:
a) Steel: -13,724,811,572
b) Brick: -5,410,155,197


### Categorías con superávit:

Plexiglass muestra la mayor diferencia positiva (5,117,731,585), lo que indica que se están vendiendo significativamente más productos de Plexiglass de los que se están exportando.
Otras categorías con diferencias positivas notables incluyen Glass, Aluminum, Plastic, Wood, Granite, y Vinyl.


### Categorías con déficit:

acero y ladrillos son las únicas categorías que muestran un déficit, pero son déficits muy significativos.

### Balance general:

El total de todas las categorías muestra un déficit de -10,230,823,465, lo que sugiere que en general, se están exportando más productos a Paraguay de los que se están vendiendo.

Posibles explicaciones:

a) Inventario: Podría haber un exceso de inventario de acero y ladrillos en Paraguay.

b) Reexportación: Es posible que algunos productos, especialmente acero y ladrillos, se estén reexportando a otros países desde Paraguay.

c) Errores de registro: Podría haber discrepancias en cómo se están registrando las ventas versus las exportaciones.

d) Timing: Las diferencias podrían deberse a diferencias en el tiempo entre cuando se registran las exportaciones y cuando se realizan las ventas.

### Categorías equilibradas:

Categorías como Rubber, Brass, y Stone muestran diferencias relativamente pequeñas, lo que sugiere un mejor equilibrio entre exportaciones y ventas.

```python
merged_df = pd.merge(df_sales[['distributor', 'Brick', 'Steel']], 
					 df_exports[['distributor', 'Brick', 'Steel']], 
					 on='distributor', 
					 suffixes=('_sales', '_exports'))

merged_df['Brick_difference'] = merged_df['Brick_sales'] - merged_df['Brick_exports']
merged_df['Steel_difference'] = merged_df['Steel_sales'] - merged_df['Steel_exports']
```

```python
brick_df = merged_df[['distributor', 'Brick_sales', 'Brick_exports', 'Brick_difference']]
print("Brick Sales and Exports:")
print(brick_df.astype(int))

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'brick_df' is your DataFrame
# If 'brick_df' is not defined, you need to create it from the provided data

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar chart for sales and exports
ax1.bar(brick_df.index - 0.2, brick_df['Brick_sales'], width=0.4, label='Brick Sales', color='blue')
ax1.bar(brick_df.index + 0.2, brick_df['Brick_exports'], width=0.4, label='Brick Exports', color='orange')

# Labels and titles
ax1.set_xlabel('Distributor')
ax1.set_ylabel('Sales/Exports')
ax1.set_title('Brick Sales vs. Exports by Distributor')
ax1.legend(loc='upper left')

# Optional: Adding Brick_difference as a line plot
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(brick_df.index, brick_df['Brick_difference'], label='Brick Difference', color='red', marker='o')
ax2.set_ylabel('Difference')
ax2.legend(loc='upper right')

# Show plot
plt.xticks(ticks=brick_df.index, labels=brick_df['distributor'].astype(str), rotation=90)
plt.tight_layout()
plt.show()
```
